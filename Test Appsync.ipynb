```python
MAX_LINE_LEN = 90
INDENTATION = ' ' * 8

def line_wrapper(line):
    ws = iter(line.strip().split(' '))
    chunk = INDENTATION + next(ws)
    for w in ws: # len(chunk) > 0
        if len(chunk) + len(w) < MAX_LINE_LEN:
            chunk += ' ' + w
        else:
            yield chunk
            chunk = INDENTATION + w
    yield chunk        
```

```python
xml_value = '\n'.join(chunk for line in cell_value.split('\n') for chunk in line_wrapper(line))
print('<string ...>', xml_value, '</string>', sep='\n')
```
